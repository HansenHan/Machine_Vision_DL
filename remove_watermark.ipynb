```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
img = cv2.imread("test2.png")

plt.imshow(img)
plt.show()
```

```python
# Solution 1:
ret, new_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(new_img)
plt.show()
```

```python
# Solution 2:
alpha = 3.0
beta = -300

new_img2 = alpha * img + beta
new_img2 = np.clip(new_img2, 0, 255).astype(np.uint8)

plt.imshow(new_img2)
plt.show()
```

```python
# I think second solution is better.
img_path = 'remove_watermark_test2.png'
plt.imsave(img_path, new_img2)
```
