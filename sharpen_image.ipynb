```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
img = cv2.imread("test1.jpg")

plt.imshow(img)
plt.show()
```

```python
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
img = cv2.filter2D(img, -1, kernel)
```

```python
plt.imshow(img)
plt.show()
```

```python
img_path = 'sharpen_test1.jpg'
plt.imsave(img_path, img)
```
